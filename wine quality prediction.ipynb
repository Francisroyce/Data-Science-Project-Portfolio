# Importing Dependencies

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

#model Evalutions
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.metrics import RocCurveDisplay
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

# Example data preprocessing
x = pd.get_dummies(x, drop_first=True)  # Convert categorical features

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# Scale the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Define models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC()
}
```

```python
def fit_and_score(models, x_train, x_test, y_train, y_test):
    np.random.seed(42)
    model_scores = {}
    for name, model in models.items():
        model.fit(x_train, y_train)  # Fit the model
        model_scores[name] = model.score(x_test, y_test)  # Evaluate and store the score
    return model_scores
```

```python
model_scores = fit_and_score(
    models=models,
    x_train=x_train_scaled,
    x_test=x_test_scaled,
    y_train=y_train,
    y_test=y_test
)

# Convert model scores to a DataFrame for easy viewing
model_scores_df = pd.DataFrame(model_scores, index=['accuracy'])

# Transpose the DataFrame to have models as rows
model_scores_df.T
```

# Confusion Matrix Analysis

This confusion matrix displays the performance of a classification model on three classes: **High**, **Low**, and **Medium** for wine quality prediction. Let's break it down:

## Key Points in the Confusion Matrix:

- **Rows:** Represent the **true labels** (actual class of the wine quality).
- **Columns:** Represent the **predicted labels** (predicted class by the model).
- **Values in Cells:** The count of predictions for each combination of true and predicted labels.

### 1. Class High:

- **True Positives (13):** The model correctly predicted 13 instances as "High."
- **False Negatives (18):** 18 instances of "High" were misclassified as "Medium."
- **False Positives (0):** No "High" was misclassified as "Low" or "Medium."

### 2. Class Low:

- **True Positives (0):** The model did not correctly classify any instance as "Low."
- **False Negatives (8):** 1 instance of "Low" was misclassified as "High," and 7 as "Medium."
- **False Positives (0):** No instances were incorrectly classified as "Low."

### 3. Class Medium:

- **True Positives (178):** The model correctly predicted 178 instances as "Medium."
- **False Negatives (12):** 12 instances of "Medium" were misclassified as "High."
- **False Positives:** Some "High" and "Low" were misclassified as "Medium."

---

## Observations:

1. **Imbalanced Predictions:**
   - The model has a strong bias toward predicting the "Medium" class. This is evident from the high number of true positives for "Medium" and the low (or zero) true positives for other classes.

2. **Poor Classification for "Low":**
   - The model completely fails to classify any instance as "Low," resulting in **0 true positives.**

3. **Misclassification of "High":**
   - While some instances of "High" are correctly predicted, a significant number (18) are misclassified as "Medium."

---

## Possible Issues and Solutions:

### 1. Class Imbalance:

- If the dataset is imbalanced (e.g., more "Medium" labels than "High" or "Low"), the model may favor the majority class.

#### Solution:

- Resample the data (oversample minority classes or undersample the majority class).
- Use class weighting during model training to penalize misclassifications of minority classes.

### 2. Model Complexity:

- The model may be too simplistic or not tuned properly, leading to poor decision boundaries between classes.

#### Solution:

- Experiment with a more complex model or fine-tune hyperparameters.
- Use techniques like feature engineering or ensemble methods.

# Loading and Preparing the Dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('WineQT.csv')

# Check for missing values
print(df.isnull().sum())

# Separate features and target
x = df.drop('quality', axis=1)  
y = df['quality']

# Encode categorical variables if needed
x = pd.get_dummies(x, drop_first=True)

# Define a function to categorize the quality ratings
def categorize_quality(rating):
    if 1 <= rating <= 4:
        return 'Low'
    elif 5 <= rating <= 6:
        return 'Medium'
    elif 7 <= rating <= 10:
        return 'High'
    else:
        return 'Unknown'  # In case there are ratings outside this range

# Create a new target column (y) based on the categorization
y = y.apply(categorize_quality)

# Check the distribution of the new target variable y
print(y.value_counts())
```

# Splitting the Dataset

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

# Handling Class Imbalance with SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42, k_neighbors=3)  # Adjust k_neighbors to less than the number of samples
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
# Define parameter grid for Random Forest
param_grid_random = {
    'n_estimators': [50, 100, 200, 500],
    'max_depth': [None, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Perform RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_grid_random,
    n_iter=50,
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

random_search.fit(x_train_resampled, y_train_resampled)
print("Best Parameters from RandomizedSearchCV:\n", random_search.best_params_)
```

```python
# Define parameter grid for fine-tuning
param_grid_fine = {
    'n_estimators': [random_search.best_params_['n_estimators'] - 50,
                     random_search.best_params_['n_estimators'],
                     random_search.best_params_['n_estimators'] + 50],
    'max_depth': [random_search.best_params_['max_depth'] - 10,
                  random_search.best_params_['max_depth'],
                  random_search.best_params_['max_depth'] + 10],
}

# Perform GridSearchCV
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid_fine,
    cv=5,
    verbose=2,
    n_jobs=-1
)

grid_search.fit(x_train_resampled, y_train_resampled)
print("Best Parameters from GridSearchCV:\n", grid_search.best_params_)
```

```python
# Train the best model
final_model = grid_search.best_estimator_
final_model.fit(x_train_resampled, y_train_resampled)

# Predictions and evaluation
y_pred_final = final_model.predict(x_test)

print("Final Classification Report:\n", classification_report(y_test, y_pred_final))
print("Final Confusion Matrix:\n", confusion_matrix(y_test, y_pred_final))
print("Final Accuracy Score:", accuracy_score(y_test, y_pred_final))
```
